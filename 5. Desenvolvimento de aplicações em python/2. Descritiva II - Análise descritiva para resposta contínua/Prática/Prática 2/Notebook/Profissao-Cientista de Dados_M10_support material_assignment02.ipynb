Considere a mesma base de dados utilizada na tarefa 1 ```previsao_de_renda.csv```.

1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

Resposta: 
- **`educação`**
    - É um ótimo indicador para fazer duas análises sobre renda. A primeira pode ser feita diferenciando homens e mulheres usando o parâmetro `hue=`, o que nos mostra que, na maioria das categorias, os homens possuem uma média de renda bem maior que a das mulheres. No entanto, pode haver distorções se analisarmos apenas pós-graduados, pois, nesse grupo, as mulheres apresentam uma média de renda superior à dos homens. Porém, essas médias são muito dispersas, com salários muito variados tanto para homens quanto para mulheres.
    - Uma segunda análise possível é manter `educação` no eixo x e diferenciar pela variável `mau` no `hue=`. Nesse caso, observamos informações relevantes sobre os bons pagadores, que possuem médias de renda mais precisas, enquanto os maus pagadores geralmente apresentam médias inferiores e menos confiáveis, sendo superiores apenas na categoria "Primário". Além disso, nos dados de pós-graduados, não há informações suficientes sobre maus pagadores, o que impede uma análise precisa para o grupo `Default=False`.
- **`estado_civil`**
    - É uma variável interessante para analisar a renda das mulheres, independente do estado civil.
    - Porém, para os viúvos, pode ocorrer uma distorção devido ao alto valor do erro padrão. Apesar de estarem na mesma média que as mulheres, o baixo nível de confiança pode gerar amostras onde a renda média dos homens seja maior ou menor que a das mulheres, dificultando a interpretação.
- **`tipo_renda`**
    - É uma boa variável para analisar a renda de homens e mulheres. Contudo, há uma limitação no caso dos bolsistas, já que só há dados de mulheres nesse grupo e, além disso, as informações apresentam baixa padronização. Isso reduz o nível de confiança nessa análise.
- **`posse_de_veículo`**
    - Quando combinada com a variável `posse_de_imóvel` no `hue=`, essa variável fornece bons indicadores sobre a renda. Embora o erro padrão seja mediano, ainda é possível analisar os dados sem gerar anomalias significativas.
- **`posse_de_imóvel`**
    - Quando analisada por sexo, essa variável oferece um ótimo panorama sobre as diferenças de renda, com um baixo erro padrão, tornando a análise mais precisa.
    - Além disso, visualizando `posse_de_imóvel` com `tipo_renda` no `hue=`, também obtemos bons resultados. Contudo, é importante desconsiderar os bolsistas novamente, já que os dados desse grupo não são confiáveis. Por outro lado, para empresários, assalariados e pensionistas, os resultados são claros e apresentam alta precisão devido ao baixo erro padrão.

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
````

````python
df = pd.read_csv('./Input/previsao_de_renda.csv')
df.drop(columns= ['Unnamed: 0', 'index'], inplace=True)
df.info()
````

````python
plt.figure(figsize=[16,5])
sns.pointplot(df, hue='posse_de_imovel',x='sexo', y='renda',dodge=True, ci = 95)
plt.show()
````
