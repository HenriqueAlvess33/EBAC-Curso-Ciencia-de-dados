####  2) Faça um boxplot do peso do bebê

```python
plt.figure()
sns_plot = sns.boxplot(data = df['PESO'])
sns_plot.set(ylabel = 'Peso do Bebê')
```

####  3) Faça um histograma da idade da mãe

```python
sns_plot = sns.displot(data = df, alpha = .25 , kde = True, element= 'step', bins = 20, x = 'IDADEMAE',)
sns_plot.set(xlabel = 'Idade da mãe', ylabel = 'Frequência')

plt.show()
```
