## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qt_acertos'])
````

````python
# 1) Seu código aqui


def calculo_da_porcentagem_de_acertos_nas_materias(dataframe, numero_de_questoes) -> str:
    porcentual_de_acerto = (dataframe['Qt_acertos'].sum()) / (numero_de_questoes*100)
    porcentual_de_acerto = round(porcentual_de_acerto * 100, 2) 
    return porcentual_de_acerto

matematica = calculo_da_porcentagem_de_acertos_nas_materias(df_mat, 24)
portugues = calculo_da_porcentagem_de_acertos_nas_materias(df_por, 18)
geografia = calculo_da_porcentagem_de_acertos_nas_materias(df_geo, 8)
ingles = calculo_da_porcentagem_de_acertos_nas_materias(df_ing, 8)
historia = calculo_da_porcentagem_de_acertos_nas_materias(df_his, 8)
fisica = calculo_da_porcentagem_de_acertos_nas_materias(df_fis, 12)
quimica = calculo_da_porcentagem_de_acertos_nas_materias(df_qui, 12)

print(f'O porcentual de acertos na matéria de matemática foi de: {matematica}%')
print(f'O porcentual de acertos na matéria de português foi de: {portugues}%')
print(f'O porcentual de acertos na matéria de geografia foi de: {geografia}%')
print(f'O porcentual de acertos na matéria de inglês foi de: {ingles}%')
print(f'O porcentual de acertos na matéria de história foi de: {historia}%')
print(f'O porcentual de acertos na matéria de física foi de: {fisica}%')
print(f'O porcentual de acertos na matéria de química foi de: {quimica}%')
````

````python
# 2) Seu código aqui
df_all = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1) 
df_all.columns = ('Matemática', 'Português', 'Geografia', 'Inglês', 'História', 'Física', 'Quimica')
dataframe_transposto = df_all.T

for nome, aluno in dataframe_transposto.items():
    total_de_acertos = aluno.sum()
    print(f'Nome: {nome}')
    print(f'{aluno.to_string(dtype=False)}')
    print(f'O total de acertos foi de: {total_de_acertos}\n')
    
df_all['Total de acertos'] = df_all.sum(axis=1)
````

````python
df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui
````

````python
# 3) Seu código aqui
total_de_questoes = {'Matemática': 24, 'Português': 18 , 'Geografia': 8, 'Inglês': 8, 'História': 8, 'Física': 12, 'Quimica': 12}
coluna_total_de_questoes = pd.Series(total_de_questoes).sum()
````

````python
df_all['Total de questões'] = coluna_total_de_questoes
df_all['Porcentagem geral de acertos'] = round((df_all['Total de acertos'] / df_all['Total de questões']) * 100, 2)
df_all
````

````python
# 4) Seu código aqui
df_all['Aprovado'] = 0
df_all.loc[df_all['Porcentagem geral de acertos'] >= 45, 'Aprovado'] = 1
df_all['Aprovado'].sum()
````
