#### 2) Vamos montar um metadados

1. Crie um dataframe com os nomes de cada variável e o tipo de dados de cada variável.
2. Adicione uma coluna nesse *dataframe* chamada "qtd_categorias" e coloque nela o número de categorias correspondente de cada variável.
    Dica:
        1. inicie uma lista vazia
        2. faça um for ao longo dos nomes das variáveis,
        3. conte o número de categorias dessa variável
        4. acumule essa informação de 3. na lista que você criou em 1.
        5. No final, essa lista pode ser atribuída à nossa variável.

````python
lista_de_colunas = []
lista_de_colunas = list(df.columns)
lista_de_colunas
````

````python
lista_de_tipos = list(df.dtypes)
lista_de_tipos = list(map(lambda x: 'object' if x == 'O' else 'int64' if x == 'int64' else 'float64' if x == 'float64' else 'bool' if x == 'bool' else x, lista_de_tipos))
````

````python
col_series = pd.Series(lista_de_colunas, lista_de_tipos)
df_coltype = pd.DataFrame(col_series)
df_coltype.reset_index(inplace=True)
df_coltype.columns = ['Tipo', 'Coluna']
df_coltype
````

````python
df['tipo_renda'].nunique()
````

````python
lista_qtd_categorias = []
df_coltype['qtd_categorias'] = 0

for index, row in df_coltype.iterrows():
   coluna = row['Coluna']
   tipo = row['Tipo']
   
   
   if tipo == 'object':
       df_coltype.loc[index,'qtd_categorias'] = df[coluna].nunique()
   else:
        df_coltype.loc[index , 'qtd_categorias'] = 0
````

````python
df_coltype
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
dicionario_dummies = {}
for index, row in df_coltype.iterrows():
    coluna = row['Coluna']
    tipo = row['Tipo']
    
    if tipo == 'object':
        series = df[coluna].value_counts()
        dado = series.index.tolist()
        dicionario_dummies[coluna] = dado
    
dicionario_dummies
````

````python
df_encoded = pd.get_dummies(df, columns= ['tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia'], drop_first=True)
df_encoded.head()
````

````python
df_encoded.dtypes
````

````python
df_encoded['sexo'].unique()
df_encoded['posse_de_veiculo'].unique()
df_encoded['posse_de_imovel'].unique()
````

````python
sexo = {'M': 1, 'F':0}
posse_de_veiculo = {'Y': 1, 'N': 0}
posse_de_imovel = {'Y': 1, 'N': 0}
````

````python

df_encoded['sexo'] = df_encoded['sexo'].map(sexo)
df_encoded['posse_de_veiculo'] = df_encoded['posse_de_veiculo'].map(posse_de_veiculo)
df_encoded['posse_de_imovel'] = df_encoded['posse_de_imovel'].map(posse_de_imovel)
df_encoded
````

````python
df_encoded.dtypes
````

````python
dicionario_dummies
````

````python
for chave, lista in dicionario_dummies.items():
    if chave not in ['sexo', 'posse_de_imovel', 'posse_de_veiculo']:
        for item in lista:
            correcao_de_item = (f'{chave}_{item}')
            if correcao_de_item in df_encoded.columns:
                df_encoded[correcao_de_item] = df_encoded[correcao_de_item].astype(int)
                                   
df_encoded.dtypes
````

````python
remanescentes = ['idade', 'tempo_emprego', 'qt_pessoas_residencia', 'mau']
for item in remanescentes:
    if item in df_encoded.columns:
        df_encoded[item] = df_encoded[item].astype(int)
        
df_encoded.dtypes
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
freq_posse_veiculo = pd.crosstab(df_encoded['mau'], df_encoded['posse_de_veiculo'])
freq_posse_veiculo
````

````python
freq_posse_email = pd.crosstab(df_encoded['mau'], df_encoded['possui_email'])
freq_posse_email
````

````python
freq = pd.crosstab(df_encoded['posse_de_veiculo'], df_encoded['possui_email'])
freq
````

````python
freq_invertida = pd.crosstab(df_encoded['possui_email'], df_encoded['posse_de_veiculo'])
freq_invertida
````

````python
freq_posse_veiculo_pct = freq_posse_veiculo.div(freq_posse_veiculo.sum(axis=1), axis=0)
freq_posse_veiculo_pct
````

````python
freq_posse_email_pct = freq_posse_email.div(freq_posse_email.sum(axis=1), axis=0)
freq_posse_email_pct
````

````python
freq_pct = freq.div(freq.sum(axis=1), axis=0)
freq_pct
````

````python
freq_invertida_pct = freq_invertida.div(freq_invertida.sum(axis=1), axis=0)
freq_invertida_pct
````

Através dos números absolutos na crosstab de possui_email e posse_de_veiculo podemos afirmar que a variável posse_de_veiculo é capaz de passar uma predição com maior precisão por sua amostragem ser maior, visto que a maior parte da concentração dos dados de possui_email está na condição de 0 e os dados são melhores representados quando posicionamos posse_de_veiculo nas colunas para que a distribuição seja feita em torno dele.
