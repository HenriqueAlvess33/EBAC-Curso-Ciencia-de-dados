####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui
def coleta_dados (uf, delegacao) :
    uf = uf.upper()
    url_estado = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Estado'
    url_municipio = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado=Munic%C3%ADpio'
    
    if delegacao == 'municipio':
        seletor = url_municipio
    elif delegacao == 'estado':
        seletor = url_estado
    else:
        print('Não foi possível realizar a busca da base de dados')
        
    
    _r = requests.get(seletor)
    
    if r.status_code == 200:
        _data_json = _r.json()
        _df = pd.DataFrame(_data_json['items'])
    else:
        print('Houve algum erro com a obtenção da Query')
        
    return _df
````

````python
coleta_dados('ba', 'municipio')
````

````python
# 2) Seu código aqui
query_minas_estado = coleta_dados('mg', 'estado')
query_minas_estado['status'].nunique()
print("É possivel encontrar 8 linha com o status de 'Arquivado por descurso de prazo'")
````

````python
# 3) Seu código aqui
query_bahia_municipio = coleta_dados('ba', 'municipio')
query_bahia_municipio.head(n=5)
query_bahia_municipio.loc[query_bahia_municipio['status'] == 'Deferido', 'interessado'].value_counts()
print("O municipio da Bahia com maior quantidade de status 'Deferido' é Luís Eduardo de Magalhães")
````

````python
# 4) Seu código aqui
query_bahia_estado = coleta_dados('ba', 'estado')
query_bahia_estado.to_csv('Busca_Bahia_Estado.csv', index=False)
pd.read_csv('Busca_Bahia_Estado.csv')
````
